```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input, decode_predictions
import numpy as np
from PIL import Image

# 1. Загружаем предобученную модель
model = MobileNetV2(weights='imagenet')

# 2. Загружаем и предобрабатываем изображение
img_path = '/content/drive/MyDrive/AI/детекция_картинки/your_image1.JPG'  # замените на путь к вашему изображению
img = Image.open(img_path).convert('RGB')
img = img.resize((224, 224))  # MobileNetV2 ожидает изображения 224x224
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# 3. Делаем предсказание
predictions = model.predict(x)

# 4. Декодируем результат (топ-3 предсказания)
decoded = decode_predictions(predictions, top=3)[0]

# 5. Выводим результат
for i, (imagenet_id, label, score) in enumerate(decoded):
    print(f"{i+1}. {label}: {score:.2f}")
```

```python
import torch
from torchvision import models, transforms
from PIL import Image
import json

# Загрузка модели
model = models.resnet50(pretrained=True)
model.eval()

# Предобработка изображения
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

img = Image.open("/content/drive/MyDrive/AI/детекция_картинки/your_image1.JPG").convert("RGB")
input_tensor = preprocess(img)
input_batch = input_tensor.unsqueeze(0)

# Предсказание
with torch.no_grad():
    output = model(input_batch)

# Загрузка меток ImageNet
with open("/content/drive/MyDrive/AI/детекция_картинки/imagenet-simple-labels.json") as f:
    labels = json.load(f)

predicted_class = output[0].argmax().item()
print("Predicted:", labels[str(predicted_class)][1])
```

```python
from transformers import ViTFeatureExtractor, ViTForImageClassification
from PIL import Image

feature_extractor = ViTFeatureExtractor.from_pretrained('google/vit-base-patch16-224')
model = ViTForImageClassification.from_pretrained('google/vit-base-patch16-224')

image = Image.open('/content/drive/MyDrive/AI/детекция_картинки/your_image1.JPG')
inputs = feature_extractor(images=image, return_tensors="pt")
outputs = model(**inputs)
logits = outputs.logits
predicted_class_idx = logits.argmax(-1).item()
print("Predicted class:", model.config.id2label[predicted_class_idx])
```
