```python
from docx import Document

def read_docx(file_path):
    doc = Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)
```

```python
import os

corpus = []
for filename in os.listdir("/content/drive/MyDrive/AI/письма/letters"):
    if filename.endswith(".docx"):
        text = read_docx(os.path.join("letters", filename))
        if text.strip():
            corpus.append(text)

# Объединим всё в один большой текст (или оставим как список)
all_text = "\n\n---НОВОЕ ПИСЬМО---\n\n".join(corpus)
```
